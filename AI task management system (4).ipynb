```python
# 1ï¸âƒ£ Import libraries
import pandas as pd
import numpy as np
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Download resources 

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')


df = pd.DataFrame(data)
print("ðŸ“Š Original Data:")
print(df.head())


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    
    text = text.lower()
    # Remove URLs and mentions
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    # Remove numbers and punctuation
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    # Remove stopwords and short tokens, then lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 2]
    # Join back
    return ' '.join(tokens)

#  Apply Preprocessing
df['clean_description'] = df['Description'].apply(clean_text)
print("\nâœ… Cleaned Text:")
print(df[['Description', 'clean_description']])

#  TF-IDF Vectorization ,that convert text to numeric features0
vectorizer = TfidfVectorizer(max_features=10, ngram_range=(1,2))
X_tfidf = vectorizer.fit_transform(df['clean_description'])

# Convert TF-IDF result to DataFrame for inspection
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
print("\nðŸ”  TF-IDF Feature Matrix:")
print(tfidf_df)

#  Combine back with original DataFrame 
final_df = pd.concat([df, tfidf_df], axis=1)

print("\nâœ… Final Preprocessed Dataset:")
print(final_df.head())
```

```python
#  Detect target column and prepare feature candidates
import pandas as pd
import numpy as np

PREFERRED_TARGETS = ['Priority', 'priority', 'Status', 'status', 'Resolution', 'resolution']
TEXT_COL_CANDIDATES = ['Description', 'description', 'Summary', 'summary']

try:
    df
except NameError:
    raise RuntimeError("DataFrame `df` not found. Load your CSV into df first, e.g. df = pd.read_csv('file.csv')")

# strip column names for safety
df.columns = df.columns.str.strip()

# Find target
target_col = None
for t in PREFERRED_TARGETS:
    if t in df.columns:
        target_col = t
        break

if target_col is None:
    # fallback: try to pick a column with few unique text-like values (categorical)
    for c in df.columns:
        if df[c].nunique() <= 10 and df[c].dtype == object:
            target_col = c
            break

if target_col is None:
    raise RuntimeError("No suitable target column auto-detected. Rename your target to 'Priority' or 'Status' or set `target_col` manually.")

print("Using target column:", target_col)
print("Target value counts:")
print(df[target_col].value_counts())

text_col = None
for cand in TEXT_COL_CANDIDATES:
    if cand in df.columns:
        text_col = cand
        break

print("Using text column (if any):", text_col)
```

```python
#  Build feature matrix X and label y
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy import sparse

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in df.select_dtypes(include=['object']).columns if c not in [target_col, text_col]]

# Keep  low-cardinality categoricals (<=30 unique)
cat_cols = [c for c in cat_cols if df[c].nunique() <= 30][:5]  # limit to 5 to avoid huge encodings

print("Numeric cols used:", num_cols)
print("Categorical cols used:", cat_cols)

if len(num_cols) > 0:
    X_num = df[num_cols].fillna(0).astype(float)
else:
    X_num = None

# Prepare X_cat (one-hot)
if cat_cols:
    ohe = OneHotEncoder(handle_unknown='ignore')# sparse=True)
    X_cat = ohe.fit_transform(df[cat_cols].fillna('NA').astype(str))
else:
    X_cat = None

if text_col:
    tfidf = TfidfVectorizer(max_features=2000, stop_words='english', ngram_range=(1,2))
    X_text = tfidf.fit_transform(df[text_col].fillna('').astype(str))
else:
    X_text = None

parts = []
if X_num is not None:
    parts.append(sparse.csr_matrix(X_num.values))
if X_cat is not None:
    parts.append(X_cat)
if X_text is not None:
    parts.append(X_text)

if not parts:
    raise RuntimeError("No features detected. Ensure dataset has numeric/categorical/text columns.")

X = sparse.hstack(parts, format='csr')
print("Final feature matrix shape:", X.shape)

# Build y (encoded)
le = LabelEncoder()
y_raw = df[target_col].fillna('UNKNOWN').astype(str)
y = le.fit_transform(y_raw)
print("Classes detected:", le.classes_)
```

```python
#  Train/test split: handle rare classes automatically
from sklearn.model_selection import train_test_split
import numpy as np

# Remove samples from classes that have < 2 occurrences 
unique, counts = np.unique(y, return_counts=True)
rare_classes = unique[counts < 2]
if rare_classes.size > 0:
    mask = ~np.isin(y, rare_classes)
    print(f"Removing {len(y) - mask.sum()} samples from rare classes: {rare_classes}")
    X = X[mask]
    y = y[mask]

# Now safe stratified split if possible
try:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
except ValueError:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=None
    )

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
# Random Forest training & evaluation
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision (macro):", precision_score(y_test, y_pred_rf, average='macro', zero_division=0))
print("Recall (macro):", recall_score(y_test, y_pred_rf, average='macro', zero_division=0))
#print(classification_report(y_test, y_pred_rf, target_names=le.classes_, zero_division=0))
#print(classification_report(y_test, y_pred_nb, zero_division=0))
print(classification_report(y_test, y_pred_rf, zero_division=0))
```
